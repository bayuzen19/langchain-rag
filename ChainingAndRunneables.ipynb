## String Parsing

```python
from langchain_core.output_parsers import StrOutputParser

prompt_template = ChatPromptTemplate(
    [
        ("system","you are master everything, answer the question in summary and bullet points with professional language."),
        ("user","please explain more details about {query}")
    ]
)
query = input("Please input your question...")
chain = prompt_template | llm | StrOutputParser()

content = chain.invoke({"query":query})
print(content)
```

### Chaining multiple Chains

```python
prompt_template = ChatPromptTemplate(
    [
        ("system","you are master everything, answer the question with professional language."),
        ("user","please explain more details about {query}")
    ]
)
query = input("Please input your question...")

#chain 1
detailedChain = prompt_template | llm | StrOutputParser()

#chain 2
headingInfoTemplate = ChatPromptTemplate.from_template("""Analyse the response and summarize in bullet points {response}""")

chainWithHeading = {"response": detailedChain } | headingInfoTemplate | llm | StrOutputParser()
content = chainWithHeading.invoke({"query":query})
print(content)
```

## Running Pararel

```python
llm = ChatOllama(
    base_url="http://localhost:11434",
    model="llama3.2:latest",
    temperature=0.5,
    max_tokens=500
)


llm2 = ChatOllama(
    base_url="http://localhost:11434",
    model="llama3.2:latest",
    temperature=0.5,
    max_tokens=500
)
```

```python
from langchain_core.runnables import RunnableParallel
from langchain_core.output_parsers import StrOutputParser
prompt_template = ChatPromptTemplate(
    [
        ("system","you are master everything, answer the question with professional language."),
        ("user","please explain more details about {query}")
    ]
)
query = input("Please input your question...")

#chain 1
detailedChain = prompt_template | llm | StrOutputParser()

#chain 2
headingInfoTemplate = ChatPromptTemplate.from_template("""Analyse the response and summarize in bullet points {response}""")

chainWithHeading = {"response": detailedChain } | headingInfoTemplate | llm2 | StrOutputParser()
pararelrunnable = RunnableParallel(chain1=detailedChain, chain2=chainWithHeading)
response = pararelrunnable.invoke({"query":query})

print(response)
```

```python
from langchain_core.runnables import RunnableParallel
from langchain_core.output_parsers import StrOutputParser
local_template = ChatPromptTemplate.from_template("please explain more details about {local}")
local = input("Please input your question 1...")
cloud = input("Please input your question 2...")
#chain 1
localChain = local_template | llm | StrOutputParser()

#chain 2
# headingInfoTemplate = ChatPromptTemplate.from_template("""Analyse the response and summarize in bullet points {response}""")

cloud_template = ChatPromptTemplate.from_template("please explain more details about {cloud}")

chainWithHeading = cloud_template | llm2 | StrOutputParser()
pararelrunnable = RunnableParallel(chain1=localChain, chain2=chainWithHeading)
response = pararelrunnable.invoke({"local":local,"cloud":cloud})

print(response["chain1"])
print("\n")
print(response["chain2"])
```
